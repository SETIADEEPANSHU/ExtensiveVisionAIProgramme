# Open CV Edge Detection Tutorial

Edge Detectors are used in Image Processing algorithms to bring out some of the patterns in the given Image

```
# import the necessary packages

import numpy as np
from urllib.request import urlopen
import cv2
from matplotlib import pyplot as plt

# METHOD #1: OpenCV, NumPy, and urllib
def url_to_image(url):
	# download the image, convert it to a NumPy array, and then read
	# it into OpenCV format
	resp = urlopen(url)
	image = np.asarray(bytearray(resp.read()), dtype="uint8")
	image = cv2.imdecode(image, cv2.IMREAD_COLOR)
 
	# return the image
	return image
```

### Importing Image

```
image = url_to_image("https://fontsarena-cd5e.kxcdn.com/wp-content/uploads/2019/04/helvetica-now-font-400x364.png")
```

### Displaying Image

```
from google.colab.patches import cv2_imshow
cv2_imshow(image)
```

### OpenCV Functions

```
edges = cv2.Canny(image,100,200)
```

```
cv2_imshow(edges)
```

### Vertical Edge Detector

```
#kernel = np.ones((3,3),np.float32)/25
kernel = np.float32([[-1,0,1],[-1,0,1],[-1,0,1]])

dst = cv2.filter2D(image,-1,kernel)
cv2_imshow(dst)
```

#Horizontal Edge Detector


```
#kernel = np.ones((3,3),np.float32)/25
kernel = np.float32([[-1,-1,-1],[0,0,0],[1,1,1]])

dst = cv2.filter2D(image,-1,kernel)
cv2_imshow(dst)
```

# 45 Degree Angle Detector


Both +ve and -ve

```
#kernel = np.ones((3,3),np.float32)/25
kernel = np.float32([[0,-1,-1],[1,0,-1],[1,1,0]])

dst = cv2.filter2D(image,-1,kernel)
cv2_imshow(dst)
```

```
#kernel = np.ones((3,3),np.float32)/25
kernel = np.float32([[0,1,1],[-1,0,1],[-1,-1,0]])

dst = cv2.filter2D(image,-1,kernel)
cv2_imshow(dst)
```

# New Image

```
image = url_to_image("https://encrypted-tbn0.gstatic.com/images?q=tbn:ANd9GcQsbil82tz99dnP7f9WSuIlu5EKiykJujMvJ0uEw6ID0GwaEnLT")
from google.colab.patches import cv2_imshow
cv2_imshow(image)
```

#Vertical Edge Detector

Detects the edges present at 90 degree

```
#kernel = np.ones((3,3),np.float32)/25
kernel = np.float32([[-1,0,1],[-1,0,1],[-1,0,1]])

dst = cv2.filter2D(image,-1,kernel)
cv2_imshow(dst)

```

#Horizontal Edge Detector

Detects the edges present at 0 degree

```
#kernel = np.ones((3,3),np.float32)/25
kernel = np.float32([[-1,-1,-1],[0,0,0],[1,1,1]])

dst = cv2.filter2D(image,-1,kernel)
cv2_imshow(dst)
```

#45 Degree Detector

Detects only edges at 45 degree

```
#kernel = np.ones((3,3),np.float32)/25
kernel = np.float32([[0,1,1],[-1,0,1],[-1,-1,0]])

dst = cv2.filter2D(image,-1,kernel)
cv2_imshow(dst)
```

#BLUR FILTER

Make the image look like blurred

```
#kernel = np.ones((3,3),np.float32)/25
kernel = 1/9 * np.float32([[1,1,1],[1,1,1],[1,1,1]])

dst = cv2.filter2D(image,-1,kernel)
cv2_imshow(dst)
```

# Sharpen Filter

Sharpens the image 

```
#kernel = np.ones((3,3),np.float32)/25
kernel = np.float32([[0,-0.5,0],[-0.5,3,-0.5],[0,-0.5,0]])

dst = cv2.filter2D(image,-1,kernel)
cv2_imshow(dst)
```

# Identity Function


Return the original Image

```
kernel = np.float32([[0,0,0],[0,1,0],[0,0,0]])

dst = cv2.filter2D(image,-1,kernel)
cv2_imshow(dst)
```
